```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt 
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, precision_score, recall_score
```

## Getting the train and test datasets, and separing them for the traning & tests procedures

```python
train = pd.read_csv('Datas/ETrain.csv')
test = pd.read_csv('Datas/ETest.csv')
y_train = train['Label']
y_test = test['Label']
X_train = train.drop(['URL','Label'],axis=1)
X_test = test.drop(['URL','Label'],axis=1)
```

## Training


### Max_depth = 2

```python
Rf2 = RandomForestClassifier(max_depth=2)
Rf2.fit(X_train,y_train)
```

```python
#Accuracy
train2accuracy = Rf2.score(X_train,y_train)
test2accuracy = Rf2.score(X_test,y_test)
#To calcule precision, recall and f1-score train
ytrain2_true = y_train
ytrain2_pred = Rf2.predict(X_train)
# To calcule precision, recall and f1-score test
ytest2_true = y_test
ytest2_pred = Rf2.predict(X_test)
#Precision - Train & Test 
train2precision = precision_score(ytrain2_true, ytrain2_pred)
test2precision = precision_score(ytest2_true, ytest2_pred)
#Recall - Train & Test
train2recall = recall_score(ytrain2_true, ytrain2_pred)
test2recall = recall_score(ytest2_true, ytest2_pred)
#F1 score - Train - Test
train2f1score = f1_score(ytrain2_true, ytrain2_pred)
test2f1score= f1_score(ytest2_true, ytest2_pred)
print('Train Accuracy: ',train2accuracy)
print('Test Accuracy: ',test2accuracy)
print('Precision train: {}, Recall train: {}'.format(train2precision,train2recall))
print('Precision test: {}, Recall test: {}'.format(test2precision,test2recall))
print('F1-score train: {}, F1-score test: {}'.format(train2f1score,test2f1score))
```

```python
plot_confusion_matrix(Rf2,X_train,y_train,cmap="Blues") #treino
```

```python
plot_confusion_matrix(Rf2,X_test,y_test,cmap="Blues") #teste
```
